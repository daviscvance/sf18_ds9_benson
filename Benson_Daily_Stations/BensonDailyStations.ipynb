# Import data (Jan 1 2017 to July 1 2017)

```python
%%time
# Source: http://web.mta.info/developers/turnstile.html
def get_data(week_nums):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [170107, 170114, 170121, 170128, 
             170204, 170211, 170218, 170225, 
             170304, 170311, 170318, 170325, 
             170401, 170408, 170415, 170422, 170429, 
             170506, 170513, 170520, 170527, 
             170603, 170610, 170617, 170624, 170701]

turnstiles_df = get_data(week_nums)

# .format fills in the {} of the url with week_nums variable
# yymmdd 170107
```

```python
turnstiles_daily2 = turnstiles_daily2.drop(["DIVISION"], axis=1, errors="ignore")
turnstiles_daily2.head()
```

```python
turnstiles_daily2[["PREV_DATE", "PREV_ENTRIES", "PREV_EXITS"]] = (turnstiles_daily2
                                                                    .groupby(["C/A", "UNIT", "SCP", "STATION", "LINENAME"])["DATE", "ENTRIES", "EXITS"]
                                                                    .transform(lambda grp: grp.shift(1)))
```

```python
turnstiles_daily2[turnstiles_daily2["ENTRIES"] < turnstiles_daily2["PREV_ENTRIES"]].head(10)
# Some turnstiles are counting backwards
```

```python
# What's the deal with counter being in reverse
mask = ((turnstiles_df["C/A"] == "A011") & 
(turnstiles_df["UNIT"] == "R080") & 
(turnstiles_df["SCP"] == "01-00-00") & 
(turnstiles_df["STATION"] == "57 ST-7 AV") &
(turnstiles_df["DATE_TIME"].dt.date == datetime.datetime(2017, 1, 1).date()))
turnstiles_df[mask].head()
```

```python
# Let's see how many stations have this problem
(turnstiles_daily2[turnstiles_daily2["ENTRIES"] < turnstiles_daily2["PREV_ENTRIES"]]
    .groupby(["C/A", "UNIT", "SCP", "STATION", "LINENAME"])
    .size())
```

```python
# Run the function twice, once for entries, and again for exits
def get_daily_counts(row, max_counter):
    counter = row["ENTRIES"] - row["PREV_ENTRIES"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        print(row["ENTRIES"], row["PREV_ENTRIES"])
        counter = min(row["ENTRIES"], row["PREV_ENTRIES"])
    if counter > max_counter:
        # Check it again to make sure we are not giving a counter that's too big
        return 0
    return counter

def get_daily_counts2(row, max_counter):
    counter = row["EXITS"] - row["PREV_EXITS"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        print(row["EXITS"], row["PREV_EXITS"])
        counter = min(row["EXITS"], row["PREV_EXITS"])
    if counter > max_counter:
        # Check it again to make sure we are not giving a counter that's too big
        return 0
    return counter

# If counter is > 1Million, then the counter might have been reset.  
# Just set it to zero as different counters have different cycle limits
```

```python
# Daily Entries
turnstiles_daily2["DAILY_ENTRIES"] = turnstiles_daily2.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
# Daily Exits
turnstiles_daily2["DAILY_EXITS"] = turnstiles_daily2.apply(get_daily_counts2, axis=1, max_counter=1000000)
```
